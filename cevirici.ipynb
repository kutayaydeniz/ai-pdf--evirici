```python
#--------------------------------
def dosyadan_oku():
    # görevler.txt varsa satırları okuyup strip ile temizlenmiş görev listesini döner
    # yoksa boş liste döner
    try:
        with open("gorevler.txt", "r", encoding="utf-8") as f:
            return [satir.strip() for satir in f.readlines()]
    except FileNotFoundError:
        return []

def dosyaya_yaz(gorevler):
    # Mevcut görev listesini gorevler.txt dosyasına yazar,kaydeder.
    with open("gorevler.txt", "w", encoding="utf-8") as f:
        for gorev in gorevler:
            f.write(gorev + "\n")

def gorevleri_listele(gorevler):
    # Görev listesini numaralı şekilde yazdırır, boşsa kullanıcıya bildirir.
    if not gorevler:
        print("Görev listesi boş.")
    else:
        print("\n--- Görevler ---")
        for i, gorev in enumerate(gorevler, 1):
            print(f"{i}. {gorev}")
    print()

def yeni_gorev_ekle(gorevler):
    # Kullanıcıdan yeni görev alır, boş değilse listeye ekler ve dosyaya yazar
    yeni = input("Yeni görev girin: ").strip()
    if yeni == "":
        print("Boş görev eklenemez!")
    else:
        gorevler.append(yeni)
        dosyaya_yaz(gorevler)
        print("Görev eklendi.")

def gorev_sil(gorevler):
    # Görev listesinden bir görev numarası seçerek silmeye yarar
    #görev silmek için mevcut liste gösterilir ve kullanıcı silmek istediği sayıyı girer
    #eğer bulunan sayılardan farklı numara veya sayı olmayan bir şey girilirse tekrar girdi istenir.
    #Uygun girdi sağlandığında görev silme işlemi biter.
    #dosyaya kaydeder.
    if not gorevler:
        print("Görev listesi boş.")
        return

    while True:
        gorevleri_listele(gorevler)
        secim = input("Silmek istediğiniz görev numarasını girin (iptal için 'q'): ").strip()
        if secim.lower() == 'q':
            print("Silme işlemi iptal edildi.")
            return
        if not secim.isdigit():
            print("Lütfen geçerli bir sayı girin!")
            continue

        silinecek = int(secim)
        if 1 <= silinecek <= len(gorevler):
            silinen = gorevler.pop(silinecek - 1)
            dosyaya_yaz(gorevler)
            print(f'"{silinen}" silindi.')
            return
        else:
            print("Geçersiz görev numarası!")

def gorev_duzenle(gorevler):
    # Seçilen görev numarasındaki metni kullanıcıdan alınan yeni metinle değiştirir
    #igili değiştirilmek istenen görev sayısı girilir,uygun girdi girilmezse
    # kullanıcıdan tekrar girilmesi istenir.uygun girdi sağlandığında
    # görev güncellenir.
    if not gorevler:
        print("Görev listesi boş.")
        return

    while True:
        gorevleri_listele(gorevler)
        secim = input("Düzenlemek istediğiniz görev numarasını girin (iptal için 'q'): ").strip()
        if secim.lower() == 'q':
            print("Düzenleme işlemi iptal edildi.")
            return
        if not secim.isdigit():
            print("Lütfen geçerli bir sayı girin!")
            continue

        duzenlenecek = int(secim)
        if 1 <= duzenlenecek <= len(gorevler):
            yeni_metin = input("Yeni görev metni: ").strip()
            if yeni_metin == "":
                print("Boş görev girilemez!")
                continue
            gorevler[duzenlenecek - 1] = yeni_metin
            dosyaya_yaz(gorevler)
            print("Görev güncellendi.")
            return
        else:
            print("Geçersiz görev numarası!")

def ana_menu():
    # Ana uygulama döngüsü: kullanıcıdan işlem seçmesini ister ve ilgili fonksiyonu çağırır
    gorevler = dosyadan_oku()
    print("Yapılacaklar Listesine Hoşgeldiniz\n")
    while True:
        secim = input(
            "Yapmak istediğiniz işlemi girin:\n"
            "1. Görevleri Listele\n"
            "2. Yeni Görev Ekle\n"
            "3. Görev Düzenle\n"
            "4. Görev Sil\n"
            "5. Çıkış\n"
            "Seçiminiz (1-5): "
        )

        if secim == "1":
            gorevleri_listele(gorevler)
        elif secim == "2":
            yeni_gorev_ekle(gorevler)
        elif secim == "3":
            gorev_duzenle(gorevler)
        elif secim == "4":
            gorev_sil(gorevler)
        elif secim == "5":
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçim. Lütfen 1-5 arası bir sayı girin.\n")

# Program başlatılır
ana_menu()

```
